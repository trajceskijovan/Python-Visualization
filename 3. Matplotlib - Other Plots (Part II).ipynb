# Matplotlib - Other Plots
> Instead of ```plt.plot(x,y)```, we can use:  
> ```plt.scatter```, ```plt.boxplot```, ```plt.pie```, ```plt.bar```, ```plt.boxplot```

````python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd

plt.style.use('ggplot')
````

## Scatterplot

````python
# Data
age = np.random.randint(20, 90, size=100)
tv  = np.random.uniform(0,50,100)

# Data Visualization
plt.scatter(age,tv)
plt.xlabel('Age')
plt.ylabel('Hours Spend on TV')
plt.title('Hours vs Age')
plt.show()
````

## Box Plots

````python
# Data
uniformSkewed = np.random.rand(100) * 100 - 40    #[-40,60)
high_outliers = np.random.rand(10) * 50 + 100     #[100,150)
low_outliers  = np.random.rand(10) * (-50) - 100  #(-150,-100]

data = np.concatenate((uniformSkewed, high_outliers, low_outliers))

# Data Visualization
plt.boxplot(data)
plt.show()
````

## Piechart

````python
# Data
values  = [12, 55, 4, 32, 14]
colors  = ['r', 'g', 'b', 'c', 'm']
explode = [0, 0, 0.2, 0, 0]  # A particular piece stands out by 0.2
labels  = ['India', 'United States', 'Russia', 'China', 'Europe']

# Data Visualization
plt.pie(values, colors= colors, labels=labels, explode = explode)
plt.title('Student Locations')
````

## Bar Chart

````python
# Data
values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']

# Data Visualization
plt.bar(range(0,5), values, color= colors)
````

## XKCD Style Chart

````python
# Clear ggplot
plt.rcdefaults()

# Data
data1 = np.ones(100)
data2 = -1*np.arange(31)
data = np.concatenate([data1,data2])

# Data Visualization
plt.xkcd()

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
#ax.spines['right'].set_color('none')
#ax.spines['top'].set_color('none')
#plt.xticks([])
#plt.yticks([])
ax.set_ylim([-30, 10])

plt.annotate(
    'THE DAY I REALIZED\nI COULD COOK BACON\nWHENEVER I WANTED',
    xy=(70, 1), arrowprops=dict(arrowstyle='->'), xytext=(15, -10))

plt.plot(data)
plt.xlabel('time')
plt.ylabel('my overall health')
````

````python
# Data (3 sets with increasing standard deviation)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Data Visualization
plt.boxplot(data, vert=True, patch_artist=True)
plt.show()
````
