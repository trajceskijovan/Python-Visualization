# Matplotlib
> **Matplotlib** is a multiplatform data visualization library built on NumPy arrays.  
> ```python
import matplotlib.pyplot as plt
%matplotlib inline
```
>  
> ```%matplotlib inline``` prints the figure in line of the codes inside the document. But it becomes default in the latest version of Jupyter.  
> To show the figure, one can enter ```plt.show()```. But this is not necessary in Jupyter.  
> Call any one of the styles using ```plt.style.use('classic')```
> ```python
import seaborn as sns
```
> provides more options for data visualization.  
> Matplotlib can generate high-quality output in a number formats, including PNG, JPG, EPS, SVG, PGF and PDF.   

> References:  
https://matplotlib.org/gallery.html  
https://matplotlib.org/3.1.0/gallery/style_sheets/style_sheets_reference.html


````python
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
plt.style.use('default')
````

````python
import numpy as np
x = np.linspace(0, 10, 100)

# Create a plot figure OBJECT
fig = plt.figure() 

plt.plot(x, np.sin(x), '-')

# Add a plot in the same figure
plt.plot(x, np.cos(x), '--')

# Set labels and titles
plt.xlabel('X Axis: Between 0 to 10')
plt.ylabel('Y Axis Values')
plt.title('Sine Cosine Curves')

# Save the figure
fig.savefig('my_figure.png')
````

````python
# Create a plot figure
plt.figure() 

# Create the first of two panels 
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x),color='blue')

# Create the second panel 
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x),color='red')
````

## Object-Oriented Interface
> Treat the figure and axes as python objects.  
> Apply methods to these objects.  

````python
# Create Figure OBJECT(empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.5, 0.5]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, np.sin(x), 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
````

````python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Axes for Large Figure
axes2 = fig.add_axes([0.2, 0.2, 0.2, 0.2]) # Axes for Small Figure

# Larger Figure Axes 1
axes1.plot(x, np.sin(x), 'b')
axes1.set_xlabel('Large: X label')
axes1.set_ylabel('Large: Y label')
axes1.set_title('Large: Title')

# Insert Figure Axes 2
axes2.plot(x, np.cos(x), 'r')
axes2.set_xlabel('Small: X label')
axes2.set_ylabel('Small: X label')
axes2.set_title('Small: Title')
````

````python
# Create a grid of plots (2 here)
# Axes will be an array of two Axes objects
fig, axes = plt.subplots(2,1)

# Call plot() method on the appropriate object
axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x))
````

## Activity 2
> 1. Generate a numpy array x  
``` 
array([0. , 0.5, 1. , 1.5, 2. , 2.5, 3. , 3.5, 4. , 4.5, 5. ])
```  
> 2. Generate a variable y which is the square of x.  
> 3. Create a panel with 1x2 boxes using ```fig, axes = plt.subplots(1,2)```.  
> 4. In box 1, plots y against x (ie. y on y-axis and x on x-axis).  
> 5. In box 2, plots x against y (ie. x on y-axis and y on x-axis).  
> 6. Name the x,y labels and titles for both boxes.  

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,5,11)
y = x**2

fig, axes = plt.subplots(1,2)

# Now use the axes object to add stuff to plot
axes[0].plot(x, y, 'r')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('title1')

axes[1].plot(y, x, 'b')
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')
axes[1].set_title('title2')
````

````python
# Subplots can provide single figure
fig, axes = plt.subplots(figsize=(12,3), dpi=200)

axes.plot(x, np.sin(x), 'r')
axes.set_xlabel('x range')
axes.set_ylabel('y values')
axes.set_title('Sin Curve')

fig.savefig("my_figure.png", dpi=200)
````

## Legend Locations
> Upper right corner: ```axes.legend(loc=1) ```  
> Upper left corner: ```axes.legend(loc=2)```  
> Lower left corner: ```axes.legend(loc=3)```  
> Lower right corner: ```axes.legend(loc=4)```  
> Optimal location: ```axes.legend(loc=0)``` 

````python
fig = plt.figure(figsize=(4,3))

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x**2, label="x**2")
axes.plot(x, x**3, label="x**3")
axes.legend(loc=0) 
````

## Line Color, Markers

````python
fig, ax = plt.subplots(figsize=(12,8))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
````

## Transparent
> ```alpha``` between 0 (transparent) and 1 (not transparent)  

````python
fig, ax = plt.subplots(figsize=(4,3))

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
````

## Axes Format

````python
# Create a plot figure
fig = plt.figure() 
plt.plot(x, np.sin(x), '-')
plt.axis([-1, 11, -1.5, 1.5])
plt.axis('tight')
````

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(x, np.sin(x), x, np.cos(x))
axes[0].set_title("Default axes ranges")

axes[1].plot(x, np.sin(x), x, np.cos(x))
axes[1].set_ylim([-2, 2])
axes[1].set_xlim([3, 5])
axes[1].set_title("Custom axes range");
````
