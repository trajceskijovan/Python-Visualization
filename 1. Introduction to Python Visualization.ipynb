# Pandas Plots
> Denote the pandas DataFrame by ```df```.  
> The built-in plot in **pandas** can be called in either format below:   
> ```python
df.plot.type
```
> ```python
df.plot(kind='type')
```
```
> where 'type' can be: line, hist, scatter, area, bar, box, pie, density, kde, etc 

````python
import numpy as np
import pandas as pd
````

## Two Sets of Data
> ```df1``` is a time-series.  
> ```df2``` is a general data set.  

````python
df1 = pd.read_csv('df1',index_col=0)
df1.head()
````

````python
df2 = pd.read_csv('df2')
df2.head()
````

## Line
> Options are in the bracket.  
> ```figsize``` controls the size of the figure (Width, Height).  
> ```lw``` controls the width of the line.  

````python
df0 = pd.Series(np.random.standard_normal(500),name='y')
df0.plot.line()
````

````python
df1.plot.line(y='A')
````

````python
df1.plot.line(y='A',figsize=(12,3),color='blue',lw=0.5)
````

````python
df1.plot.line(y=['A','B'],figsize=(12,3),color=['blue','red'],lw=0.5)
````

## Histogram
> ```bins``` defines the number of bins.  

````python
df1['A'].plot.hist()
````

````python
df1['A'].hist(bins=30,color='blue')
````

````python
df1['A'].plot(kind='hist',bins=30,color='green')
````

## Plotting Styles
> The styles (style: ggplot, bmh, dark_background, etc) can be controlled by 
```python
import matplotlib.pyplot as plt
plt.style.use('style')
```
> **Reference**: https://matplotlib.org/3.1.0/gallery/style_sheets/style_sheets_reference.html


````python
import matplotlib.pyplot as plt
plt.rcdefaults() 
plt.style.use('default')
df1['A'].plot.hist(figsize=(4,2))
````

````python
plt.rcdefaults() 
plt.style.use('ggplot')
df1['A'].plot.hist(figsize=(4,2))
````

````python
plt.rcdefaults() 
plt.style.use('bmh')
df1['A'].plot.hist(figsize=(4,2))
````

````python
plt.rcdefaults() 
plt.style.use('Solarize_Light2')
df1['A'].plot.hist(figsize=(4,2))
````

````python
plt.rcdefaults() 
plt.style.use('classic')
df1['A'].plot.hist(figsize=(4,2))
````

````python
plt.rcdefaults() 
plt.style.use('fast')
df1['A'].plot.hist(figsize=(4,2))
````

## Area Plot 
> Let's convert back to ggplot style.  
> ```alpha``` controls the degree of transparency.  

````python
plt.rcdefaults() 
plt.style.use('ggplot')
````

````python
df2.plot.area(alpha=0.4)
````

## Bar Plots
> ```stacked=True``` generates single bar for each x value.  
> ```stacked=False``` generates a figure with multiple bars for each x value.  

````python
df2.plot.bar()
````

````python
df2.plot.bar(stacked=True)
````

## Scatterplots
> (x,y) defines your x and y variables.  
> ```c``` generates dependency between values and **color**.  
> ```cmap``` controls the color.  
> ```s``` generates dependency between values and **size**. 

````python
df2.plot.scatter(x='a',y='b',figsize=(3,3))
````

````python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
````

````python
df1.plot.scatter(x='A',y='B', s=df1['C']*200)
````

## Box Plot 

````python
df2.plot.box() # Can also pass a by= argument for groupby
````

## Hexagon

````python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
````

# Activity 1
> Generate 10000 numbers that are N(0,1).  
> Create a histogram with 25 bins, red in color.  

````python
import numpy as np
import pandas as pd
x  = np.random.normal(0,1,10000)
df = pd.DataFrame(x,columns=['x1'])
df['x1'].plot.hist(bins=25,color='red')
````

# Activity 2
> Set ```y=2*x+u``` where ```u``` is also N(0,1).  
> Create a scatterplot with y on y-axis and x on x-axis.  

````python
u  = np.random.normal(0,1,10000)
y  = 2*x+u

df = pd.DataFrame({'x1':x,'y1':y})
df.plot.scatter(x='x1',y='y1')

p   = np.linspace(-4,4,101)
pp  = 2*p

df1 = pd.DataFrame(pp,index=p,columns=['y2'])
df1['y2'].plot.line(color='red')
````
